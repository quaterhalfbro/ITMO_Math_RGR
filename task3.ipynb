**Сформулируйте признаки сравнения для определения сходимости несобственных интегралов.**

* $\int_a^b{f(x)}{\rm d}x$ сходится, если существует $g(x): f(x) <= g(x)$ на [a, b]
* $\int_a^b{f(x)}{\rm d}x$ расходится, если существует $g(x): f(x) >= g(x)$ на [a, b]

**Оцените сверху и снизу трансцендентную функцию (логарифм или арктангенс) для сравнения исходного интеграла с интегралом вида $\int_a^b{\frac{1}{x^\beta}}{\rm d}x$.Установите, при каких значениях параметра это сравнение позволяет сделать вывод о сходимости интеграла.**

Если $0 < ln(x) <= 1$, т.е. $1 < x <= e$, то $\frac{1}{x^a * ln(x)} >= \frac{1}{x^{\beta}} \to$ для $a > 1$, $\int_a^b{\frac{1}{x^\beta}}{\rm d}x$ - расходится.

Если $ln(x) >= 1$, т.е. $x >= e$, то $\frac{1}{x^a * ln(x)} <= \frac{1}{x^{\beta}} \to$ для $0 < a < 1$, $\int_a^b{\frac{1}{x^\beta}}{\rm d}x$ - сходится.
